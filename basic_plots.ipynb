```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
campaign_df = pd.read_csv("Resources/campaign.csv")
campaign_df.head()
```

```python

# Outcome of Campaigns Bar Chart
plt.figure(figsize=(8, 5))
outcome_counts = campaign_df['outcome'].value_counts()
outcome_counts.plot(kind='bar')
plt.title('Outcome of Campaigns')
plt.xlabel('Outcome')
plt.ylabel('Number of Campaigns')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("outcome_barchart.png")
plt.show()
```

```python
#Pledged vs. Goal
#A scatter plot to visualize the relationship between the goal of each campaign and the amount pledged. 
#This can help identify if there's a trend between the campaign goals and the success in meeting those goals.

# Plotting
plt.figure(figsize=(8, 5))
#sns.scatterplot(data=campaign_df, x='goal', y='pledged', alpha=0.6)
plt.scatter(campaign_df["goal"],campaign_df["pledged"])
plt.title('Campaign Goal vs. Pledged Amount')
plt.xlabel('Goal ($)')
plt.ylabel('Pledged Amount ($)')
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.savefig("goal_pledged.png")
plt.show()
```
